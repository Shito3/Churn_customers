#### Definition and issue of attrition

###### Definition
*  Attrition (or churn) is a marketing concept characterized by the abandonment of a brand, product or service by existing customers. It is calculated as a rate. This is referred to as customer churn rate. It corresponds to the level of loss of customers relative to the overall customer base over a given period of time.
* The churn rate customer is the exact opposite of the retention rate. If you add these two percentages, you get 100% because attrition measures lost customers while retention (or retention) measures the customers kept. The customer loss ratio must therefore be integrated into all strategies aimed at retaining its buyers.

## Calculating and interpreting the attrition rate
### Attrition rate formula
* The attrition rate is usually calculated monthly so that you can follow up regularly and visualize any positive or negative changes quickly.
  To calculate the attrition ratio,it is necessary to group the following data:
    •         Number of customers at the beginning of month M
    •         Number of customers lost in month M
  The formula to apply, once the data above is collected, is as follows:
  Churn misses - (Number of customers lost over the period / Total number of customers at the beginning of the period) x 100
  For example, if a brand finds that it has only 450 subscribers to its product X on December 31, whereas it had 500 subscribers on thefirst December, it recorded an attrition rate of 10% for the month of December.
  Churn misses X - [500-450) / 500] x 100.

Result analysis
There is no repository to say that the churn rate customer is satisfactory if it is between this and that value. It all depends on the sector of activity of the company concerned and the typology of its clientele. However, you have to have the idea that the lower this percentage, the more it means that the company manages to retain its buyers. A high retention rate is of course very beneficial for any organization.


##### Possible causes of customer attrition

    Customer attrition can have three causes:
    •         The abandonment of a branded product for another item of the same sign (then called relative attrition)
    •         Abandoning the product or service in favour of a competing property
    •         Abandonment without replacement, which consists of no longer using the product or service in question at all (for example, an individual who abandons his car in favour of public transport)
            The last two cases represent a total attrition for society. It is the latter that will have to be avoided above all in order not to undermine the profitability of the company.
            These causes of attrition are the result of a change in the consumer's perception and a change in the consumer's purchasing behaviour. It is then necessary to identify the external and internal factors of the firm that may influence this churn.
